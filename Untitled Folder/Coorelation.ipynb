### Pearson Correlation

```python
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
```

```python
sb.pairplot(cars)
```

```python
X = cars[['mpg','cyl','qsec','hp']]
sb.pairplot(X)
```

### Calculating pearson correlation coefficent  using scipy

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coefficient, p_value = pearsonr(mpg,hp)
print('pearson correlation coefficient  %0.3f '%(pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,qsec)
print('pearson correlation coefficient  %0.3f '%(pearsonr_coefficient))
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg,wt)
print('pearson correlation coefficient  %0.3f '%(pearsonr_coefficient))
```
