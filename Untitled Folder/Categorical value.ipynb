```python
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head(15)
```

```python
carb = cars.carb
carb.value_counts()
```

```python
cars_cat = cars[['cyl','mpg','vs','carb','gear']]
cars_cat.head(15)
```

```python
gear_set = cars.groupby('gear')
gear_set.describe()
```

### Transforming variable into categorical data types

```python
cars['group'] = pd.Series(cars.cyl,dtype='category')
cars['group'].dtypes
```

```python
cars['group'].value_counts()
```

### Describing categorical values with crosstabs

```python
pd.crosstab(cars['am'],cars['cyl'])
```
